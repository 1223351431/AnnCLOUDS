```python
def plot_variable_pie(data_table,variable):
    #准备绘图数据
    data =data_table[variable].value_counts()
    #创建饼图
    fig1,ax1= plt.subplots()
    ax1.pie(data,labels=data.index,autopct='%1.1f%%')
    plt.axis('equal')
    
    plt.title('Pie Chart of'+variable)
    plt.show()
```

```python
df1 = pd.read_csv(R'C:\Users\Administrator\Desktop\AnnCLOUDS\data\类别变量练习数据.csv',encoding= 'GBK')
```

```python
plot_variable_pie(df1,'父亲职业重新编码')
```
